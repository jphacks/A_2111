```
!pip install facenet_pytorch
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
import numpy as np
import torchvision
```

```
import torch
resnet = InceptionResnetV1(pretrained='vggface2').eval()
dummy_input = torch.tensor(np.array(Image.open("sample.jpg"))).reshape(1,3,160,160).float()
print(dummy_input.shape)
```

```
print(resnet(dummy_input).shape)
```

```
torch.onnx.export(resnet, dummy_input, "model2.onnx",input_names=["input"], output_names=["ooutput"])
```

```
ort_sess = ort.InferenceSession('model2.onnx')
```

```
outputs = ort_sess.run(None, {'input': dummy_input.detach().cpu().numpy()})
```

```
len(outputs[0][0])
```

```
def run(url):
    x=np.array(Image.open(url)).reshape(1,3,160,160).astype(np.float32)
    ort_sess = ort.InferenceSession('model2.onnx')
    outputs = ort_sess.run(None, {'input': x})
    return outputs[0][0]
run("sample.jpg")
```
